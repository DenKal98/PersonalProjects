```python
import pandas as pd 
import numpy as np
from xgboost import XGBClassifier
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
```

```python
X = pd.read_csv('train.csv').dropna()
Y = pd.read_csv('test.csv').dropna()
```

```python
Y = Y['dep_var']
```

```python
X = X.drop(['title_home','title_improv','title_payoff','title_pay'],axis=1)
```

```python
correlated_features = set()
correlation_matrix = X.corr()

for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > 0.8:
            colname = correlation_matrix.columns[i]
            correlated_features.add(colname)
```

```python
X = X.drop(correlated_features,axis=1)
```

```python
X = X[['annual_income','credit_card_balance','credit_card_usage','credit_score','debt_to_income','delinq_2yrs_cat_no','employment_length',
       'home_ownership_MORTGAGE','home_ownership_OWN','inq_last_6mths_cat_no','installment'
       ,'interest_rate','issue_date_Dec','nr_accounts','open_accounts','pub_rec_cat_no','purpose_credit_card','purpose_small_business','term_ 36 months','year']]
```

```python
sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X, Y)
X_train, X_test, y_train, y_test = train_test_split(X_train_res,y_train_res, test_size=0.3, random_state=100)
```

```python
parameters = {
    'eta': np.linspace(0, 1, 15),
    'subsample' : np.linspace(0, 1, 15) 
             }
xgb_c = XGBClassifier()
clf = GridSearchCV(xgb_c, parameters , cv=5, verbose=5, n_jobs=1)
clf.fit(X_train, y_train.ravel())
```
